```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
x1 = np.arange(1,11,3)
```

```python
x
```

```python
y = x**3
```

```python
plt.plot(x,y)
```

```python
plt.bar(x,y)
```

```python
plt.hist(x,y)
plt.title('Histogram')
plt.show()

plt.scatter(x,y)
plt.title('Scatter')
plt.show()

plt.box(x,y)
plt.title('BOx')
plt.show()

plt.violinplot(x,y)
plt.title('Violinplot')
plt.show()
```

```python
plt.box(x,y)
```

```python
y2 = x**4
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y=x**3','y2=x**4'])
plt.show()
```

```python
fig = plt.figure()
plt.plot(x,y, lw=3,marker='o',markersize=10,markerfacecolor='yellow')
plt.plot(x,y2, marker ='v',markersize = 10)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```
